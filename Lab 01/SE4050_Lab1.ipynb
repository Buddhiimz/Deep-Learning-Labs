```python
#Question 3

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')

# 7. Set labels and title
ax.set_title("3D plot of Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
#Question 4

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
# Display first 5 observations
df.head()
```

```python
# Select relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_selected = df[features]

# Compute Pearson correlation matrix
pearson_corr = df_selected.corr(method='pearson')

# Compute Spearman correlation matrix
spearman_corr = df_selected.corr(method='spearman')

# Set plot size
plt.figure(figsize=(12, 5))

# Plot Pearson correlation heatmap
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation')

# Plot Spearman correlation heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Correlation')

plt.tight_layout()
plt.show()
```

```python
#Task 2
#Question 1

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/DL Lab1 Dataset/Au_nanoparticle_dataset.csv")

# Filter only the required columns
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
new_df = df[selected_columns]

# View the new dataframe
new_df.head()
```

```python
#Question 2

# View the new dataframe
new_df.head(20)
```

```python
#Question 3

# Mean
print("Mean:\n", new_df.mean())

# Standard deviation
print("\nStandard Deviation:\n", new_df.std())

# Quartiles (25%, 50%, 75%)
quartiles = new_df.describe().loc[['25%', '50%', '75%']]
print("\nQuartiles:\n", quartiles)
```

```python
#Question 4

# 1x4 layout of histograms
plt.figure(figsize=(20, 5))

for i, col in enumerate(selected_columns):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[col], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'{col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Question 5

# Pairplot (basic scatter plots + histograms)
sns.pairplot(new_df)
```

```python
#Question 6

# Assuming 'new_df' contains only: N_total, N_bulk, N_surface, R_avg

# Create the customized pair grid
g = sns.PairGrid(new_df)

# Upper triangle: bivariate histograms
g.map_upper(sns.histplot)

# Diagonal: histogram + KDE
g.map_diag(sns.histplot, kde=True)

# Lower triangle: 2D KDE plots
g.map_lower(sns.kdeplot, fill=True)

# Optional: Display the plot
plt.show()
```
